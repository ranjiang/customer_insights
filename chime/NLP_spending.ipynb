```python
#combine the three sources
import os
from google.colab import files

# Paths to your uploaded files (update names if needed)
files = {
    "google_play": "chime_g_reviews.csv",         # uploaded manually
    "ios": "chime_ios_reviews.csv",
    "reddit": "chime_reddit_praw.csv"
}

dataframes = []

# Load and standardize each source
for platform, path in files.items():
    if os.path.exists(path):
        df = pd.read_csv(path)
        df["platform"] = platform

        # Rename and normalize text columns
        if "content" in df.columns:  # Google Play
            df = df.rename(columns={"content": "text", "score": "rating"})
        elif "review" in df.columns:  # iOS
            df = df.rename(columns={"review": "text"})
        elif "text" in df.columns:
            df["text"] = df["text"].astype(str)

        # Add missing columns if needed
        if "rating" not in df.columns:
            df["rating"] = None
        if "topic" not in df.columns:
            df["topic"] = None

        # Keep only relevant columns
        df = df[["text", "rating", "platform", "topic"]]
        dataframes.append(df)
    else:
        print(f"âš ï¸ File not found: {path}")

# Combine all sources
if dataframes:
    df_combined = pd.concat(dataframes, ignore_index=True)
    df_combined.to_csv("chime_reviews_combined.csv", index=False)
    print("âœ… Combined data saved as chime_reviews_combined.csv")
    df_combined.head()
else:
    print("âŒ No files were loaded. Please check your filenames and upload them.")
```

```python
# Run BERTopic on datesets
import pandas as pd
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

# Load your combined file
df = pd.read_csv("chime_reviews_combined.csv")

# Drop missing or short texts
df = df[df["text"].notnull()]
df = df[df["text"].str.len() > 20]

# Run BERTopic
topic_model = BERTopic(verbose=True)
topics, probs = topic_model.fit_transform(df["text"].tolist())

# Add topics back to the DataFrame
df["topic"] = topics
df["topic_name"] = df["topic"].map(topic_model.get_topic_info().set_index("Topic")["Name"])

# Save results
df.to_csv("chime_reviews_with_topics.csv", index=False)
topic_model.save("chime_bertopic_model")

# Show top topics
topic_model.get_topic_info().head(10)
```
